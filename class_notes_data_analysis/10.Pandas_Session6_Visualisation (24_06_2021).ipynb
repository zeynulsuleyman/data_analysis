```python
x = np.arange(1,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, label="y=x")
ax.plot(x, z, label="y=2x")
ax.plot(x, t, label="y=log(x)")
ax.set_xlabel("x")
ax.set_ylabel("y or z")
ax.legend(fontsize=16)
ax.set_title("...");
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline  
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
# functional olarak (fonksiyon çağırarak. fonksiyonun içine değişkeni/değişkenleri yazarak)

plt.plot(salary)  # plt.plot(x , y)
```

```python
# functional

plt.plot(age, salary)

# şimdi age değişkenleri ile birleşti.
```

```python
# iki çizgi çıksın istersem:
# ikinci çizgi için farklı bir değişken (salary_2) kullanarak y eksinini kaydırıyorum

plt.plot(age, salary)
plt.plot(age, salary_2)  # ikinci plotta y eksinine farklı değer girdiğim için ikinci çizgiyi çizgi
```

```python
# legend kullanırsak:
# legend'de label ları ya plt.plot() fonksiyonunun içinde tanımlıyoruz.
 
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.legend()
plt.show()
```

```python
# veya legend() fonksiyonunun içinde de tanımlayabiliriz.
 
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.legend(["mavi", "turuncu"])
```

```python
#plt.plot(age, salary)
#plt.plot(age, salary_2)
plt.plot(age, salary, age, salary_2) 
plt.legend(["1", "2"])

# iki farklı axes'ı (age-salary ve age-salary_2) sırayla fonksiyon içine soktum. iki çizgili şekil elde ettim.
```

```python
# fonksiyonda 2 axes kullandığımda legend'in label'larını fonksiyon içinde değil, legend() fonksiyonunun içinde tanımlıyorum!!!

plt.plot(age, salary, age, salary_2)

plt.legend(["1", "2"]);
```

```python
# object oriented olarak:

# önce object'leri oluşturuyoruz ( tanımlıyoruz )

# sonra .plot metodu ile verilerimi giriyoruz
# plt.plot(x,x, label="..")
fig, abc = plt.subplots()
abc.plot(age, salary_2)
plt.legend(["age-sal2"]);
```

```python
# 1. figure size ını genişleterek düzeltebiliriz.

plt.figure(figsize=(4,3))

plt.subplot(1,2,1)
plt.plot(age, salary)

plt.subplot(1,2,2)
plt.plot(age, salary_2)
```

```python
# 2. plt.tight_layout() ile düzelttik.

plt.figure(figsize=(4,3))
plt.subplot(2,2,1)
plt.plot(age, salary)

plt.subplot(2,2,2)
plt.plot(age, salary_2)

plt.subplot(2,2,3)
plt.plot(age, salary_2)

plt.subplot(2,2,4)
plt.plot(age, salary_2)

plt.tight_layout()
```

```python
# !!!! Hüseyin hocam ve Raife hocam

def multi_graf(rows, cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):
            ax[i,j].plot(age, salary)
            plt.tight_layout()
multi_graf(2,2)
```

# Box Plot

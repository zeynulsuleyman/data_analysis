```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary);
```

```python
plt.plot(salary);
```

```python
# Funtional method
plt.plot(age, salary)

plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
# OOP method

fig, ax = plt.subplots()  # figure is the top level container for all the plot elements
ax.plot(age,salary, 'r')  # axes represents an individual plot
ax.set_xlabel("age")
ax.set_ylabel('salary')
ax.set_title("Salary by Age")
plt.show()
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
x = np.arange(1, 11)
y = x.copy()
z = x * 2
t = np.log(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, z)
ax.plot(x, t)
ax.set_xlabel("x")
ax.set_ylabel("t, y or z")
ax.set_title("...");
```

```python
x = np.arange(1,11)
y = x.copy()
z = x*2
t = np.log(x)

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, y, label="y=x")
ax.plot(x, z, label="y=2x")
ax.plot(x, t, label="y=log(x)")
ax.set_xlabel("x")
ax.set_ylabel("y or z")
ax.legend(fontsize=16)
ax.set_title("...");
```
